```python
#Reading data file and setting variables

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder 

dataTrain = pd.read_csv("/kaggle/input/home-data-for-ml-course/train.csv") # Main data being used to train the model (not to be confused with X_train and y_train)
dataTest = pd.read_csv("/kaggle/input/home-data-for-ml-course/test.csv") # THIS IS THE DATA THAT THE MODEL WILL USE AT THE END FOR FINAL SUBMISSION.

X = dataTrain.iloc[:, 1:80]
y = dataTrain.iloc[:, -1] #refer to https://www.geeksforgeeks.org/pandas/select-rows-columns-by-name-or-index-in-pandas-dataframe-using-loc-iloc/ if review on this notation is needed
encoder = LabelEncoder()

for column in X.columns:
    if X[column].dtype=="object": # go through each column and check if it is of type string
        X[column] = encoder.fit_transform(X[column]) # if the column is of type string, convert each type of string to an integer
X = X.fillna(X.mean(numeric_only=True)) # Correctly troubleshoots NaN problem (replaces all NaN values with the column mean after encoding all values)
        


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=42)



print(X.shape)
print(y.shape)
print(X.columns)
print(X[X.columns[0]].dtype=="object")
print(X[X.columns[1]].dtype=="object") # used to print true, now prints false. Indicates that transformer is working
print(X)
np.any(np.isnan(X))
```

```python
# Creating the model and training it

from sklearn.neighbors import KNeighborsRegressor

knn_regressor =  KNeighborsRegressor(n_neighbors=4)
knn_regressor.fit(X_train, y_train)
# for review on knn refer to https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html and https://www.geeksforgeeks.org/machine-learning/k-nearest-neighbors-knn-regression-with-scikit-learn/
```

```python
# Accuracy Function

def compute_accuracy(predictions, answers):
    correct=0
    for i in range(0,len(predictions)):
        if predictions[i]==answers[i]:
            correct=correct+1
    accuracy=round((correct/len(predictions))*100,2)
    return accuracy
```

```python
from sklearn.metrics import mean_squared_error
import math


y_pred = knn_regressor.predict(X_test)
RMSE = math.sqrt(mean_squared_error(y_test, y_pred))
print(RMSE)
```
